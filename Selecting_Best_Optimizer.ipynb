#Import libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Read Dataset
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Churn_Modelling.csv')
df
```

# Data Transformation

```python
df.drop(columns = ['RowNumber',	'CustomerId',	'Surname'], inplace = True)
```

```python
df
```

```python
df = pd.get_dummies(data = df, columns = ['Geography','Gender'])
df
```

# Feature Scaling

```python
x = df.drop(columns = 'Exited')
y = df['Exited']
```

```python
x
```

```python
sc = StandardScaler()

x = sc.fit_transform(x)
```

# Model Building

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.25, random_state = 42)
```

```python
ann = Sequential()

ann.add(Dense(units = 10, activation = 'relu'))

ann.add(Dense(units = 1, activation = 'sigmoid'))

ann.compile(optimizer = 'adadelta', loss = 'binary_crossentropy', metrics = 'accuracy')

ann.fit(xtrain,ytrain, epochs = 100)
```

```python
ypred = ann.predict(xtest)
ypred
```

```python
#Evaluate the model
```

```python
ypred = ypred > 0.5
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```

```python
import keras_tuner as kt
```

```python
def build_model(hp):
  model = Sequential()
  model.add(Dense(units = 10, activation ='relu'))
  model.add(Dense(units = 1, activation = 'sigmoid'))
  opti = hp.Choice('optimizer', values = ['adam','sgd','rmsprop'])
  model.compile(optimizer = opti, loss = 'binary_crossentropy', metrics = 'accuracy')

  return model
```

```python
tuner = kt.RandomSearch(build_model,
                        objective ='val_accuracy',
                        max_trials = 5,
                        directory = 'project',
                        project_name = 'ANN')
```

```python
#New model
```

```python
tuner.search(xtrain,ytrain, epochs = 5, validation_data = (xtest,ytest))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
x.shape
```

```python
model = tuner.get_best_models(num_models = 1)[0]
model.build(x.shape)
```
